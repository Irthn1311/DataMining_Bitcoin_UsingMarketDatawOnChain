# Data preprocessing

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_path = "../data/btc_onchain_daily.csv"
```

```python
data = pd.read_csv(data_path)

data
```

## Duplication data

```python
data_dup = data[data.duplicated(keep='first')]

data_dup
```

## Missing data

```python
data_missing_count = data.isnull().sum()
print(data_missing_count)
```

```python
data_with_na = data[data.isna().any(axis=1)]

data_with_na
```

- Drop na column

```python
data_dropna = data.dropna()

data_dropna.shape
```

- Round the values to 0 decimal (like integer)

```python
float_cols = data_dropna.select_dtypes(include=['float64']).columns
print(float_cols)

data_dropna[float_cols] = data_dropna[float_cols].round(0).astype('int64')

data_dropna.head(5)
```

## Save preprocessed data

```python
pre_data= "../data/preprocessed/btc_onchain_daily_preprocessed.csv"
data_dropna.to_csv(pre_data, index=False)
```
